```python
#Create a figure that showcases the yearly average price per dozen of eggs by Ontario Farmers 

#obs for this to work --> need to install openpyxl onto dsi_participant env
#conda install -n dsi_participant openpyxl

#set up env
import pandas as pd
import matplotlib.pyplot as plt

# Path to your Excel file
file_path = "/Users/lilianawu/visualization/02_activities/assignments/eggs_in_shell.xlsx"

# Read Excel file
df = pd.read_excel(file_path, header=1)

# Preview the data
print(df.head())
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['$/dozen'], marker='o', linestyle='-', color='#4a7abc', linewidth=2)
plt.title("Average Prices Received by Ontario Farmers for Eggs in Shell")
plt.xlabel("Year")
plt.ylabel("Price ($ per dozen)")
plt.grid(True)
plt.show()
```
