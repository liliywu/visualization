```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
#Before
fig, ax = plt.subplots(figsize=(5, 3))

#now we want to have two plots next to each other,
fig,(ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3))
```

```python
# two different figures!
fig, (ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
fig.show()
```

```python
#No grid arrangement
fig, someaxes = plt.subplot_mosaic(
    [['ax1', 'ax3'],
     ['ax2', 'ax3']],
    figsize=(7, 4)
)

for label, ax in someaxes.items():
    ax.text(0.5, 0.5, f'This is {label!r}',
            fontsize=14,
            ha='center',
            transform=ax.transAxes)

plt.show()
```

```python
#add data to each subplot

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
```

```python
#large x-axis 
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel(' The Cow',fontsize=18)
someaxes["ax2"].set_xlabel(' GOES MOOOH',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
plt.show()
```

```python
#constrained layout
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4),
layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
plt.show()
```

```python
# plot overlayed on plot
# first make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])
# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))
# now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1,
color = "indigo")
ax.plot(x, y2,
color = "red")
```

```python
#ADD error information

#sd of the data
y2_sd = np.std(y2)

#then plot the line as before
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

#add error bar
ax.errorbar(x, #our x values
y2, #our y values
yerr = y2_sd,
fmt = "none")
```

```python
#change the error bar appearance

fig, ax = plt.subplots(figsize=(7, 3))
# Plot the line
ax.plot(x, y2, color="red")

# Add error bars
ax.errorbar(x, y2,
            yerr=y2_sd,
            fmt="none",
            ecolor="indigo",
            elinewidth=4,
            capsize=6,
            capthick=4)
```

```python
#    USING IMAGES FROM INTERNET!

#SET UP 
from PIL import Image # to open images
import requests # to get images from URLs
from io import BytesIO # to store images
```

```python
#get images form the internet
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax_image = fig.add_axes([0.1, # x coordinate (ON FIGURE, NOT AXES)
0.11, # y coordinate (ON FIGURE, NOT AXES)
0.15, # image width
0.35]) # image height
```

```python
#Luffy! lol
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2,
color = "red")
ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])
ax_image.imshow(image)
ax_image.axis('off')
plt.show()
```
