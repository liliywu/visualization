```python
#Set up 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)


fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
```

```python
#Adding a legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right')
fig.show()
```

```python
#modify the legends --> make it horizontal
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
frameon = True, #add frame around the legend
fontsize = 12, #change font size
ncol = 2, #specify number of columns
shadow = True )
fig.show()
```

```python
#Make the legend to situate outside of the figure (upper left)
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left',
bbox_to_anchor =(1, 1))
fig.show()
```

```python
#transform the dataset into a scatterplot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

#then specify the coordinates!

ax.text(10, 95, "This value is important!",
ha='center', ##alignment of text
color = 'red', ##modify font colour
size = 20) ##modify font size
fig.show()
```

```python
#make an empty plot with the following axis 
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

#then plot! using text
ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData )
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes )
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure )
```

```python
#use annotate function to focus on specific points
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
arrowprops=dict(facecolor='black', shrink=0.05))
fig.show()

#how to change the arrow style :D  make it hot pink
ax.annotate('This is important!'
,
xy=(10, 95), xytext=(20, 94),
arrowprops = dict(arrowstyle = "wedge",
color = "hotpink"))
fig.show()
```

```python
#remove any ticks and marks 
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
#define maximum number of ticks 
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
#use a 5 increments interval for the ticks :D
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
#rotate axis labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation=45, ha='right')
```

```python
#what about modifying the title?
font1 = {'family':'serif','color':'indigo'}
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict = font1)
```
