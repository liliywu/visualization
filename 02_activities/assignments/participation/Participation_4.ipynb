```python
#install and load seaborn 
#popular package --> similar to text files 

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#load data
tips = sns.load_dataset("tips")
print(tips)
```

```python
#make a simple line plot 
sns.lineplot(data=tips, # choose our dataset
x='total_bill', # define our x variable
y='tip') # define our y variable
```

```python
#alter the appearence of the plot 
sns.set_style('whitegrid')

#add labels
tipgraph = sns.lineplot(data=tips,
x='total_bill',
y='tip')
tipgraph.set(title='Tips vs. Total Bill',
xlabel='Total Bill ($)',
ylabel='Tip Amount ($)')
```

```python
#change figure size 
fig = plt.subplots(figsize=(10, 3))

#change color + style
tipgraph = sns.lineplot(data=tips,
x='total_bill',
y='tip',
color = 'hotpink',
linestyle = '--',
linewidth = 3,
marker = 'o',
markerfacecolor = 'indigo')
```

```python
# Create the scatterplot
tipgraph = sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    style='time',
    hue='day',
    palette=['purple', 'hotpink', 'deepskyblue', 'yellowgreen']
)

# Set title and labels
tipgraph.set(
    title='Tips vs. Total Bill',
    xlabel='Total Bill ($)',
    ylabel='Tip Amount ($)'
)
```

```python
#compare variables using pairplot
sns.pairplot(data = tips,hue = 'day')
```

```python
#replot --> compare specific levels of variables
daysplot = sns.relplot(data=tips,
                       x="total_bill",
                       y="tip",
                       hue="sex",
                       col="day",
                       kind="scatter",
                       col_wrap=2)
```

```python
#import packages!
import plotly.graph_objects as go # 'go' is 'graph objects
x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
#make figure
graph = go.Figure()
```

```python
#choose plot type
graph.add_trace(go.Bar(x=x1, y=y1))

#save the plot as HTML
graph.write_html("pirategraph.html")
```

```python
#update layout
graph.update_layout(
title="Pirate Scores",
xaxis_title="Pirates",
yaxis_title="Score")
graph.show()
```

```python
#customize
graph = go.Figure()
graph.add_trace(go.Scatter(x=x1, y=y1, mode='markers', # we want points for a scatter plot
marker=dict(
size=15, # point size
color='hotpink', # point colour
opacity=1, # point transparency/alpha
line=dict(width=5, color='purple') # point outline
)))
graph.update_layout(
title='Interactive Pirate Plot',
xaxis_title='Pirates',
yaxis_title='Scores',
width=500, height=500)
```

```python
#import word cloud 
from wordcloud import WordCloud
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
                 on_bad_lines='skip')
df
```

```python
##make a simple word cloud using the quote variable

# join all our text from each row from our quote column into a string
text = " ".join(each for each in df.quote)
# generate our wordcloud image
wordcloud = WordCloud(background_color="white",
colormap = 'inferno').generate(text)
# use matplotlib syntax to put our image in a figure
fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud, # remember 'imshow' from when we added pictures to our matplotlib axes
interpolation='bilinear') # this line helps smooth our image
ax.axis("off")
plt.show()
```

```python
#make venn viagrams!
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

#define sets
A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])


#then assign sets  to each circle 
diagram = venn2_unweighted([A, B],
                           set_labels = ('Fruits', 'Pies'),
                           set_colors=("blue", "red"),
                           alpha=0.5)

#add some labels 
diagram.get_label_by_id("10").set_text("\n".join(A - B))
diagram.get_label_by_id("11").set_text("\n".join(A & B))
diagram.get_label_by_id("01").set_text("\n".join(B - A))
plt.show()
```
