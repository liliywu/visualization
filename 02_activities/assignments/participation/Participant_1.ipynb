```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
import PIL 
import requests 
```

```python
 np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
#Modify our graph 
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
#line plot!
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
#make a histogram!
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
#add some labels
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()

#visualize
fig
```

```python
#modify appearance of labels 
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}
#visualize
fig
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
#modifying with loc argument
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
# star plots! haha cute
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x,
y,
marker='*', # NEW ADDITIONS
color = "indigo") # NEW ADDITIONS
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,
y,
marker='*',
color = 'indigo',
linestyle = '--', # NEW ADDITIONS
linewidth = 2) # NEW ADDITIONS
fig.show()
```

```python
#change some colors!
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker='*', color='#7425b6', linestyle='--', linewidth=2)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x, y,
        marker='*',
        markersize=12,
        color='#7425b9',
        linestyle='--',
        linewidth=2,
        markeredgecolor='#fa9359',  # NEW ADDITIONS
        markerfacecolor='#fa9359')  # NEW ADDITIONS

ax.grid(axis = 'y'
, color = "blue", linewidth = 2, linestyle = '-.')

```

```python
# ACTIVITY!

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
sns.set_style("white")
sns.kdeplot(x=df.sepal_width, y=df.sepal_length)
plt.show()
```

```python
#change colors!
sns.set_style("white")
sns.kdeplot(x=df.sepal_width, y=df.sepal_length, cmap="Reds", fill=True)
plt.show()
```

```python
#change smoothing
sns.set_style("white")
sns.kdeplot(x=df.sepal_width, y=df.sepal_length,
            cmap="Blues", fill=True, bw_adjust=0.5)
plt.show()
```
