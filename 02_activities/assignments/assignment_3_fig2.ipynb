```python
#Create a figure 

# Re-import necessary libraries after environment reset
import pandas as pd
import matplotlib.pyplot as plt

# Path to the uploaded file
file_path = "/Users/lilianawu/visualization/02_activities/assignments/cucumber_en.xlsx"

#load the data
df_clean = pd.read_excel(file_path)

# Preview the cleaned dataframe
df_clean.head(), df_clean.columns
```

```python
#clean white spaces
df_clean.columns = df_clean.columns.str.strip()

# Convert necessary columns to numeric
df_clean['Year'] = pd.to_numeric(df_clean['Year'])
df_clean['Harvested Area      (acres)'] = pd.to_numeric(df_clean['Harvested Area      (acres)'])
df_clean['Average Price   (cents/lb)'] = pd.to_numeric(df_clean['Average Price   (cents/lb)'])
df_clean['Average Yield      (lbs/acre)'] = pd.to_numeric(df_clean['Average Yield      (lbs/acre)'])

# Create subplots (3 stacked)
fig, axs = plt.subplots(3, 1, figsize=(12, 10), sharex=True)

#  Harvested Area
axs[0].plot(df_clean['Year'], df_clean['Harvested Area      (acres)'],
            marker='s', color='green', linestyle='-', linewidth=2)
axs[0].set_ylabel('Harvested Area (acres)')
axs[0].grid(True, linestyle='--', alpha=0.6)
axs[0].set_title("Cucumber and Gherkin Production Metrics in Ontario")

# Average Price
axs[1].plot(df_clean['Year'], df_clean['Average Price   (cents/lb)'],
            marker='o', color='blue', linestyle='-', linewidth=2)
axs[1].set_ylabel('Avg Price (cents/lb)')
axs[1].grid(True, linestyle='--', alpha=0.6)

#  Average Yield
axs[2].plot(df_clean['Year'], df_clean['Average Yield      (lbs/acre)'],
            marker='^', color='red', linestyle='-', linewidth=2)
axs[2].set_ylabel('Avg Yield (lbs/acre)')
axs[2].set_xlabel('Year')
axs[2].grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
